```python
import pandas as pd
import random
from faker import Faker
```

```python
fake = Faker()
customers = [] #this initializes fake data generator
```

```python
for i in range(1,10001):
    customers.append([
        i,
        fake.name(),
        random.choice(['Male','Female']),
        random.randint(18,70),
        fake.city(),
        fake.state(),
        fake.date_between(start_date='-5y',end_date='today')
    ])
        
#Creates 10,000 customers
#Each customer has:
#ID,Name,Gender,Age,City,State,Join Date
```

```python
customers_df = pd.DataFrame(customers, columns=[
    'customer_id',
    'customer_name',
    'gender',
    'age',
    'city',
    'state',
    'join_date'
])
#This converts data into table form.
```

```python
customers_df.head()
```

```python
customers_df.head(10)
```

```python
customers_df.to_csv("customers.csv", index=False)
```

```python
import random
import pandas as pd
import random
from faker import Faker

fake = Faker()
product = []
categories = {
    'Electronics':['Mobile','Laptop','TV'],
    'Fashion': ['Shirt', 'Shoes', 'Jeans'],
    'Grocery': ['Snacks', 'Beverages', 'Dairy'],
    'Home': ['Furniture', 'Decor', 'Kitchen']
}
brands = ['Samsung', 'Nike', 'Apple', 'Sony', 'Adidas', 'LG', 'Puma']
for i in range(1, 501):
    category = random.choice(list(categories.keys()))
    sub_category = random.choice(categories[category])

    cost = random.randint(100, 5000)
    selling = cost + random.randint(50, 2000)

    product.append([
        i,
        fake.word().capitalize(),
        category,
        sub_category,
        random.choice(brands),
        cost,
        selling
    ])
    
```

```python
product_df = pd.DataFrame(product, columns=[
    'product_id',
    'product_name',
    'category',
    'sub_category',
    'brand',
    'cost_price',
    'selling_price'
])
```

```python
product_df.head()
```

```python
product_df.to_csv("product.csv", index=False)
```

```python
import random
import pandas as pd
from faker import Faker

fake = Faker()
stores = []

for i in range(1, 51):
    stores.append([
        i,
        f"Store_{i}",
        fake.city(),
        fake.state(),
        fake.name()
    ])
```

```python
stores_df = pd.DataFrame(stores, columns=[
    'store_id',
    'store_name',
    'city',
    'state',
    'manager_name'
])
```

```python
stores_df.head()
```

```python
stores_df.to_csv("stores.csv", index=False)
```

```python
import random
import pandas as pd
from faker import Faker
orders = []

payment_methods = ['Cash', 'Card', 'UPI', 'Net Banking']

for i in range(1, 50001):
    orders.append([
        i,
        random.randint(1, 10000),   # customer_id
        random.randint(1, 50),      # store_id
        fake.date_between(start_date='-2y', end_date='today'),
        random.choice(payment_methods)
    ])
```

```python
orders_df = pd.DataFrame(orders, columns=[
    'order_id',
    'customer_id',
    'store_id',
    'order_date',
    'payment_method'
])
```

```python
orders_df.head()
```

```python
orders_df.to_csv("orders.csv", index=False)
```

```python
import random
import pandas as pd
from faker import Faker
order_details = []

for i in range(1, 100001):
    order_details.append([
        i,
        random.randint(1, 50000),   # order_id
        random.randint(1, 500),     # product_id
        random.randint(1, 10),      # quantity
        round(random.uniform(0, 30), 2)   # discount %
    ])
```

```python
order_details_df = pd.DataFrame(order_details, columns=[
    'order_detail_id',
    'order_id',
    'product_id',
    'quantity',
    'discount'
])
```

```python
order_details_df.head()
```

```python
order_details_df.to_csv("order_details.csv", index=False)
```
